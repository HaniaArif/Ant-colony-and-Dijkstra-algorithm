```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng


#This class is used for colors in the command prompt

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[36m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

arraySizeX = 0
arraySizeY = 0

#Asking user the size of maze they want (Along then down)

print("How long is the width of your array?")
arraySizeX = int(input())


print("How long is the height of your array?")
arraySizeY = int(input())


rng = default_rng()
randomArray = rng.integers(0, 10 , (arraySizeY,arraySizeX), )
currentElement = 0

#Defining a function which prints out the array with a chosen element to be highlighted

def printArray(arrayx,arrayy,array,currentElement):
    for x1 in range(arrayx):
        print("")
        for y1 in range(arrayy):
            if(x1 == currentElement//arrayx and y1 == currentElement % arrayx):
                print("\033[91m" + str(array[x1][y1])  + "\033[0m" , end = ' ')
            else:
                print(array[x1][y1], end = ' ')

#Printing array

printArray(arraySizeY,arraySizeX,randomArray,currentElement)



```

# Nearest neighbour (only in the direction of right and down)

```python
def neighbor():
    path = []
    currentNode = 0
    length = 0
    while (currentNode != (arraySizeX*arraySizeY - 1)):
        s = abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode+arraySizeX)//arraySizeX][(currentNode+arraySizeX) % arraySizeX]))
        d = abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode + 1)//arraySizeX][(currentNode + 1) % arraySizeX]))
        if (currentNode%arraySizeX < (arraySizeX - 1) and currentNode//arraySizeX < (arraySizeY - 1)):
            if (s < d):
                currentNode += arraySizeX
                length += s
            else:
                currentNode += 1
                length += d
        elif(currentNode%arraySizeX < (arraySizeX - 1)):
            currentNode += 1
            length += d
        else:
            currentNode += arraySizeX
            length += s
        path.append(currentNode)
    path.append(currentNode)
    print(path)

    for y1 in range(arraySizeY):
            print("")
            for x1 in range(arraySizeX):
                check = 0
                for scan in range(len(path)-1):
                    if(path[scan] == x1+(y1*arraySizeX)):
                        print("\033[91m" + str(randomArray[y1][x1])  + "\033[0m" , end = ' ')
                        check = 1
                if(check == 0):
                    print(str(randomArray[y1][x1]) , end = ' ')
    
    print("\nThe length: ", length)
print(arraySizeX*arraySizeY - 1)
neighbor()
    
```

# Dijkstra

```python
#Dijkstra's algorithm

def dijkstra():

    currentNode = 0
    map = np.zeros((arraySizeX*arraySizeY,2,1))

    map[0][1] = 1

    def checkState(map):
        for x1 in range(0,arraySizeX*arraySizeY):
                if (map[x1][1] == 0):
                    return False    
        return True

    def findNextNode(map):
        min = 10000000000
        findingNode = 0
        for x in range(arraySizeX*arraySizeY):
            if (map[x][1] == 1):
                try:

                    if (map[x+1][1] == 0 and map[x+1][0] <= min):
                        min = map[x+1][0]
                        findingNode = x+1
                except:
                    pass
                try:
                    if (map[x-1][1] == 0 and map[x-1][0] <= min):
                        min = map[x-1][0]
                        findingNode = x-1
                except:
                    pass
                try:

                    if (map[x + arraySizeX][1] == 0 and map[x+ arraySizeX][0] <= min):
                        min = map[x + arraySizeX][0]
                        findingNode = x + arraySizeX
                except:
                    pass
                try:

                    if (map[x - arraySizeX][1] == 0 and map[x - arraySizeX][0] <= min):
                        min = map[x - arraySizeX][0]
                        findingNode = x - arraySizeX
                except:
                    pass
        return findingNode

    currentNode = 0

    def calculateD(direction,currentNode):
        if(direction == "w"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode-arraySizeX)//arraySizeX][(currentNode-arraySizeX) % arraySizeX]))
        elif(direction == "s"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode+arraySizeX)//arraySizeX][(currentNode+arraySizeX) % arraySizeX]))
        elif(direction == "a"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode - 1)//arraySizeX][(currentNode - 1) % arraySizeX]))
        elif(direction == "d"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode + 1)//arraySizeX][(currentNode + 1) % arraySizeX]))


    for change in range(1,arraySizeX*arraySizeY):
        map[change][0] = 999

    while (checkState(map) != True):
        map[currentNode][1] = 1
        if (currentNode >=arraySizeX):
            if (map[currentNode - arraySizeX][1] != 1 and (map[currentNode][0] + calculateD("w",currentNode) < map[currentNode - arraySizeX][0])):
                map[currentNode - arraySizeX][0] = map[currentNode][0] + calculateD("w",currentNode)
        if (currentNode//arraySizeX < (arraySizeY - 1)):
            if (map[currentNode + arraySizeX][1] != 1 and (map[currentNode][0] + calculateD("s",currentNode) < map[currentNode + arraySizeX][0])):
                map[currentNode + arraySizeX][0] = map[currentNode][0] + calculateD("s",currentNode)

        if (currentNode%arraySizeX > 0):
            if (map[currentNode - 1][1] != 1 and (map[currentNode][0] + calculateD("a",currentNode) < map[currentNode - 1][0])):
                map[currentNode - 1][0] = map[currentNode][0] + calculateD("a",currentNode)
        
        if (currentNode%arraySizeX < (arraySizeX - 1)):
            if (map[currentNode + 1][1] != 1 and (map[currentNode][0] + calculateD("d",currentNode) < map[currentNode + 1][0])):
                map[currentNode + 1][0] = map[currentNode][0] + calculateD("d",currentNode)

        currentNode = findNextNode(map)

    def checkPop(number, poparray):
        for x in poparray:
            if(x == number):
                return True
        return False

    def printSolution(arrayx,arrayy,map):
        for x1 in range(0,arrayy):
            print("")
            for y1 in range(0,arrayx):
                print(str(map[y1+(x1*arrayx)][0]) , end = ' ')

        traversal = arrayx*arrayy - 1
        nodesThatWePassed = [arrayx*arrayy - 1]

        lastDirection = 1
        popList = []
        lastDirection1 = ""
        while (traversal != 0):
            try:
                if (abs(map[traversal][0] - map[traversal - 1][0]) == calculateD("a",traversal) and map[traversal][0] >= map[traversal - 1][0] and traversal %arrayx > 0 and checkPop(traversal-1,popList) == False and lastDirection1 != "d"):
                    traversal = traversal - 1
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "a"
                    lastDirection += 1
                    if (lastDirection == arrayx*arrayy/2):
                        break
                elif (abs(map[traversal][0] - map[traversal - arrayx][0]) == calculateD("w",traversal) and map[traversal][0] >= map[traversal - arrayx][0] and traversal >=arrayx and checkPop(traversal- arrayx,popList) == False and lastDirection1 != "s"):
                    traversal -= arrayx
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "w"
                elif (abs(map[traversal][0] - map[traversal + 1][0]) == calculateD("d",traversal) and map[traversal][0] >= map[traversal + 1][0] and (traversal % arrayx) < arrayx - 1 and checkPop(traversal + 1,popList) == False and lastDirection1 != "a"):
                    traversal = traversal + 1
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "d"
                elif (abs(map[traversal][0] - map[traversal + arrayx][0]) == calculateD("s",traversal) and map[traversal][0] >= map[traversal + arrayx][0] and (traversal // arrayx) < arrayy -1 and checkPop(traversal + arrayx,popList) == False and lastDirection1 != "w"):
                    traversal = traversal + arrayx
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "s"
                else:
                    popList.append(nodesThatWePassed.pop(len(nodesThatWePassed) - 1))             
                    traversal = nodesThatWePassed[len(nodesThatWePassed) - 1]
            except:
                    break


        print("\n")
        print(list(reversed(nodesThatWePassed)))

        for y1 in range(arrayy):
            print("")
            for x1 in range(arrayx):
                check = 0
                for scan in range(len(nodesThatWePassed)-1):
                    if(nodesThatWePassed[scan] == x1+(y1*arrayx)):
                        print("\033[91m" + str(randomArray[y1][x1])  + "\033[0m" , end = ' ')
                        check = 1
                if(check == 0):
                    print(str(randomArray[y1][x1]) , end = ' ')
        

    print("\n")
    printSolution(arraySizeX,arraySizeY,map)
```
