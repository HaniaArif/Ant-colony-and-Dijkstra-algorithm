# Nearest neighbour (only in the direction of right and down)

```python
import timeit
def neighbor():

    #Creating variables for the nearest neighbor.
    tic = timeit.default_timer()
    path = []
    currentNode = 0
    length = 0

    #While loop stops when the function has reached the destination (the bottom right node).

    while (currentNode != (arraySizeX*arraySizeY - 1)):
        #Calculating the absolute difference between the node we are current on and the adjacent bottom and right node.
        #The position of the current node can be represent as a number,
        #Where the number represents the position of the node as if the array was single vector.
        #The value can be converted to the position in the 2d array by using div and mod respectively.
        #We will be using this throughout all 3 solutions.


        #The adjacent bottom node is the size of the "x axis" away from the current node if represented as a single vector.
        #Therefore we can represent the node below as currentNode+arraySizeX
        #And the node to the right is currentNode+1

        #s and d represent the direction (wasd)

        s = abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode+arraySizeX)//arraySizeX][(currentNode+arraySizeX) % arraySizeX]))
        d = abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode + 1)//arraySizeX][(currentNode + 1) % arraySizeX]))

        #If statement is used to see if the node is on the 'edge' of the array.

        if (currentNode%arraySizeX < (arraySizeX - 1) and currentNode//arraySizeX < (arraySizeY - 1)):
            if (s < d):
                currentNode += arraySizeX
                length += s
            else:
                currentNode += 1
                length += d

        #A problem with this solution is that it priorities going right than down if the two nodes are the same number.
        #This could potentially affect the outcome of the final length.

        #These next two else if statements are used if the current node are on the 'edge' of the array.

        elif(currentNode%arraySizeX < (arraySizeX - 1)):
            currentNode += 1
            length += d

        else:
            currentNode += arraySizeX
            length += s

        #Record the nodes we have passed through.

        path.append(currentNode)
    

    path.append(currentNode)
    print(path)


    #This is used to print out the solution of the problem and is not part of the algorithm.
    #Mainly used to show a clearer solution by adding colour.

    for y1 in range(arraySizeY):
            print("")
            for x1 in range(arraySizeX):
                check = 0
                for scan in range(len(path)-1):
                    if(path[scan] == x1+(y1*arraySizeX)):
                        print("\033[91m" + str(randomArray[y1][x1])  + "\033[0m" , end = ' ')
                        check = 1
                if(check == 0):
                    print(str(randomArray[y1][x1]) , end = ' ')
    
    # Print out the final length of the solution
    print("\nThe length: ", length)
    toc = timeit.default_timer()
    print("\nThe time to compute: ", toc - tic)
    
```

# Dijkstra

```python
import timeit
#Dijkstra's algorithm
def dijkstra():
    tic = timeit.default_timer()
    currentNode = 0

    #Creating a mapping system for each node in the matrix.
    #First value represents the shortest length currently to that node from the start.
    #Second value tells us if we have found the shorest length where 1 represents true.

    map = np.zeros((arraySizeX*arraySizeY,2,1))

    #The starting node is already at the shortest length (0) so mark it 1 (true).

    map[0][1] = 1

    #Before we start the algorithm, we are making the current shortest length in the map equal to a really high number which will be eventually overwritten.

    for change in range(1,arraySizeX*arraySizeY):
        map[change][0] = 999

    #This function checks if we have found the shortest route for all the nodes in the matrix.
    #returns false if no.

    def checkState(map):
        for x1 in range(0,arraySizeX*arraySizeY):
                if (map[x1][1] == 0):
                    return False    
        return True

    #After finding the shortest route for one node, we need to find the next node to work on.

    def findNextNode(map):
        min = 10000000000
        findingNode = 0
        currentNode = 0
        #For loop to check every node.

        for x in range(arraySizeX*arraySizeY):
            
            #If statements to find the next node that has the current shortest route marked and with no true marked as well.
            #The try and catch statements are used if the node we are looking for is out of bounds, so we ignore it.

            if (map[x][1] == 1):

                try:
                    if (map[x+1][1] == 0 and map[x+1][0] <= min):
                        min = map[x+1][0]
                        findingNode = x+1
                except:
                    pass

                try:
                    if (map[x-1][1] == 0 and map[x-1][0] <= min):
                        min = map[x-1][0]
                        findingNode = x-1
                except:
                    pass

                try:
                    if (map[x + arraySizeX][1] == 0 and map[x+ arraySizeX][0] <= min):
                        min = map[x + arraySizeX][0]
                        findingNode = x + arraySizeX
                except:
                    pass

                try:
                    if (map[x - arraySizeX][1] == 0 and map[x - arraySizeX][0] <= min):
                        min = map[x - arraySizeX][0]
                        findingNode = x - arraySizeX
                except:
                    pass
        
        #return the node we have found.
        return findingNode

    #This function is used to calculate the difference between the current node and the adjacent nodes.
    #Directions are represented by "wasd" which are top, left, down, right respectively.
    #Using something similar to the nearest neighbor algorithm, we are converting a value that represents a value in a vector to a value in an array.
    #Using div and mod.
    #Overall used to reduce repeated code as we use these many times.

    def calculateD(direction,currentNode):

        if(direction == "w"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode-arraySizeX)//arraySizeX][(currentNode-arraySizeX) % arraySizeX]))
        elif(direction == "s"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode+arraySizeX)//arraySizeX][(currentNode+arraySizeX) % arraySizeX]))
        elif(direction == "a"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode - 1)//arraySizeX][(currentNode - 1) % arraySizeX]))
        elif(direction == "d"):
            return abs((randomArray[currentNode//arraySizeX][currentNode % arraySizeX]) - (randomArray[(currentNode + 1)//arraySizeX][(currentNode + 1) % arraySizeX]))

    #While loop to check if we have found the shortest route for all the nodes in the matrix.

    while (checkState(map) != True):

        #Current node is marked 1 as we have found the shortest path to it.

        map[currentNode][1] = 1

        #Updating any adjacent nodes if we have found a shorter route to it than the current one.

        if (currentNode >=arraySizeX):
            if (map[currentNode - arraySizeX][1] != 1 and (map[currentNode][0] + calculateD("w",currentNode) < map[currentNode - arraySizeX][0])):
                map[currentNode - arraySizeX][0] = map[currentNode][0] + calculateD("w",currentNode)
        if (currentNode//arraySizeX < (arraySizeY - 1)):
            if (map[currentNode + arraySizeX][1] != 1 and (map[currentNode][0] + calculateD("s",currentNode) < map[currentNode + arraySizeX][0])):
                map[currentNode + arraySizeX][0] = map[currentNode][0] + calculateD("s",currentNode)

        if (currentNode%arraySizeX > 0):
            if (map[currentNode - 1][1] != 1 and (map[currentNode][0] + calculateD("a",currentNode) < map[currentNode - 1][0])):
                map[currentNode - 1][0] = map[currentNode][0] + calculateD("a",currentNode)
        
        if (currentNode%arraySizeX < (arraySizeX - 1)):
            if (map[currentNode + 1][1] != 1 and (map[currentNode][0] + calculateD("d",currentNode) < map[currentNode + 1][0])):
                map[currentNode + 1][0] = map[currentNode][0] + calculateD("d",currentNode)

        #Find the next shortest node marked that hasn't been marked.

        currentNode = findNextNode(map)

    #Everything below is used to find the path to take when going from the start to finish.

    #Sometimes when finding the path, we go the wrong way because two adjacent values could be the same so we need to go backwards.
    #Therefore we need to pop values and record that we have visited that value.
    #The recording makes sure we never go there again.

    def checkPop(number, poparray):
        for x in poparray:
            if(x == number):
                return True
        return False

    #This functions find the path from the end to the beginning of the matrix as dijksras works backwards after finding the shortest distance from start to finish.  

    def printSolution(arrayx,arrayy,map):

        #Here for fun we print out the shortest distance possible from each node to the beginning
        #This isn't part of the algorithm.

        print("Shortest distance from beginning to each node.")

        for x1 in range(0,arrayy):
            print("")
            for y1 in range(0,arrayx):
                print(str(map[y1+(x1*arrayx)][0]) , end = ' ')

        #Here we mark the currentNode we are on as 'traversal' as we are going backwards to the beginning.
        #nodesThatWeHave passed records the nodes that we have passed.

        traversal = arrayx*arrayy - 1
        nodesThatWePassed = [arrayx*arrayy - 1]
        popList = []
        lastDirection1 = ""

        #While that is stopped when we have reached the beginning of the matrix.

        while (traversal != 0):
            try:

                #Here we are using a lot of conditions to make sure we reach the end with the most optimal route.

                #First of all, we are making sure that the different between the adjacent nodes are the same as the difference between the adjacent nodes in the 'map'
                #This assures us that we are going the right way.

                #Secondly, we making sure the condition above is true when the current node we are on is bigger than the adjacent node on the 'map'.

                #Thirdly, we need to make sure that we are not going of the edge of the matrix.

                #Then we check if we have went to a node that we have already visited.
                #The visited node happens because the if statements below priorities going left,up,down then right.
                #There are cases when the adjacent nodes are the same values, so we need to go backwards just in case we are going the wrong way.

                #Lastly, we cant go back to a node that we just departed from.

                #With all these conditions, we are able to find the most optimal route from the beginning to end.

                if (abs(map[traversal][0] - map[traversal - 1][0]) == calculateD("a",traversal) and map[traversal][0] >= map[traversal - 1][0] \
                    and traversal %arrayx > 0 and checkPop(traversal-1,popList) == False and lastDirection1 != "d"):
                    
                    #below every if statement we are adding the node to the list and updating the current node.

                    traversal = traversal - 1
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "a"

                elif (abs(map[traversal][0] - map[traversal - arrayx][0]) == calculateD("w",traversal) and map[traversal][0] >= map[traversal - arrayx][0] \
                    and traversal >=arrayx and checkPop(traversal- arrayx,popList) == False and lastDirection1 != "s"):

                    traversal -= arrayx
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "w"

                elif (abs(map[traversal][0] - map[traversal + 1][0]) == calculateD("d",traversal) and map[traversal][0] >= map[traversal + 1][0] \
                    and (traversal % arrayx) < arrayx - 1 and checkPop(traversal + 1,popList) == False and lastDirection1 != "a"):

                    traversal = traversal + 1
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "d"

                elif (abs(map[traversal][0] - map[traversal + arrayx][0]) == calculateD("s",traversal) and map[traversal][0] >= map[traversal + arrayx][0] \
                    and (traversal // arrayx) < arrayy -1 and checkPop(traversal + arrayx,popList) == False and lastDirection1 != "w"):

                    traversal = traversal + arrayx
                    nodesThatWePassed.append(traversal)
                    lastDirection1 = "s"

                else:
                    
                    #This else statement is used if the first condition is not met.
                    #We go back in time to try again on a different node.

                    popList.append(nodesThatWePassed.pop(len(nodesThatWePassed) - 1))             
                    traversal = nodesThatWePassed[len(nodesThatWePassed) - 1]
            except:
                    break


        print("\n")

        #The list we have created was from end to beginning so we reverse the list to get beginning to end.

        print(list(reversed(nodesThatWePassed)))

        #This is not part of the algorithm, just to print out the solution with colours.

        for y1 in range(arrayy):
            print("")
            for x1 in range(arrayx):
                check = 0
                for scan in range(len(nodesThatWePassed)-1):
                    if(nodesThatWePassed[scan] == x1+(y1*arrayx)):
                        print("\033[91m" + str(randomArray[y1][x1])  + "\033[0m" , end = ' ')
                        check = 1
                if(check == 0):
                    print(str(randomArray[y1][x1]) , end = ' ')
        
    #Call the functions within the function
    #Tic tok used to time how long the processing time was.

    print("\n")
    printSolution(arraySizeX,arraySizeY,map)
    toc = timeit.default_timer()
    print(toc - tic)
```

# Ant Colony Optimization

```python
import timeit
from random import random
from random import randint

#For ant colony optimization, we are expanding the matrix to allow space for the pheromones which will be place inbetween the nodes.
#For example, a 3x3 matrix will turn into a 5x5 matrix

#1 - 2 - 3
#| x | x |
#4 - 5 - 6
#| x | x |
#7 - 8 - 9

#The numbers are the nodes, the lines are the pheromones and the x is blank spaces.

#Class used to make ants.
#The ants will move a random direction base on probability until they reach the end.

class Ants:

    #Each ant starts at the beginning of the matrix.
    #Similar with the other two algorithms, we will be converting an integer value into a space in the 2d array.

    def __init__(self):
        self.currentNode = 0
        self.lastDirection = "Null"

    #This function returns a number which will be added to the current node of the ant depending on the direction it want to go to.

    def updateCurrentNode(self,character,arraySizeX):
        if character == "w":
            return (-2*(arraySizeX*2-1))
        elif character == "a":
            return -2
        elif character == "s":
            return (2*(arraySizeX*2-1))
        elif character == "d":
            return +2
        return 0

    #Similar to dijkstras, we are calculating the difference between the current node and adjacent node next to it.
    #It is very similar except for the fact that we have expanded our matrix to an even bigger one, so we multiply somethings by 2 when calculating.
    #This gets it own function as we use it multiple times.

    def calculateDifference(self, character,trail):

        if character ==  "w":
            return (trail[self.ModandDiv("//")][self.ModandDiv("%")]) - (trail[(self.currentNode-2*(arraySizeX*2-1))//(arraySizeX*2-1)][(self.currentNode-2*(arraySizeX*2-1)) % (arraySizeX*2-1)])
        elif character == "a":
            return (trail[self.ModandDiv("//")][self.ModandDiv("%")]) - (trail[(self.currentNode - 2)//(arraySizeX*2-1)][(self.currentNode - 2) % (arraySizeX*2-1)])
        elif character == "s":
            return (trail[self.ModandDiv("//")][self.ModandDiv("%")]) - (trail[(self.currentNode + (2 * (arraySizeX*2-1)))//(arraySizeX*2-1)][(self.currentNode+(2*(arraySizeX*2-1))) % (arraySizeX*2-1)])
        elif character == "d":
            return (trail[self.ModandDiv("//")][self.ModandDiv("%")]) - (trail[(self.currentNode + 2)//(arraySizeX*2-1)][(self.currentNode + 2) % (arraySizeX*2-1)])


    #This function returns the adjacent pheromone depending on the direction given.

    def returnPheromone(self, character,trail):
        if character == "s":
            return trail[(self.currentNode+(arraySizeX*2-1))//(arraySizeX*2-1)][(self.currentNode+(arraySizeX*2-1))% (arraySizeX*2-1)]
        elif character == "d":
            return trail[(self.currentNode+1)//(arraySizeX*2-1)][(self.currentNode+1)% (arraySizeX*2-1)]
        elif character == "w":
            return trail[(self.currentNode-(arraySizeX*2-1))//(arraySizeX*2-1)][((self.currentNode)-(arraySizeX*2-1))% (arraySizeX*2-1)]
        elif character == "a":
            return trail[(self.currentNode-1)//(arraySizeX*2-1)][(self.currentNode-1)% (arraySizeX*2-1)]
        return 0

    #This used to reduce line space.

    def ModandDiv(self, symbol):
        if symbol == "//":
            return self.currentNode//(arraySizeX*2-1)
        elif symbol == "%":
            return self.currentNode % (arraySizeX*2-1)

    #This function is used to record the pheromones that ant has passed.
    #Only recording the pheromone between the nodes once.
    #Else we ignore it.

    #We also update the last direction and current node that the ant was on.

    def moveAnt(self,i,arraySizeX,trail):
        self.lastDirection = i
        if i == "w":
            adjNode = self.currentNode - arraySizeX*2-1
        elif i == "a":
            adjNode = self.currentNode - 1
        elif i == "s":
            adjNode = self.currentNode + arraySizeX*2-1
        elif i == "d":
            adjNode = self.currentNode + 1
        if adjNode not in self.uniqueNodes:
            self.totalLength += abs(self.calculateDifference(i,trail))
            self.uniqueNodes.append(adjNode)
        self.currentNode += self.updateCurrentNode(i,arraySizeX)


    #This function moves the ant base on probability.
    #First we calculate the denominator by using the sum of the pheromone multiplied by 1/distance between the nodes.
    #Then roll a random number between 0 to 1 which will land between the probablities of going to a node to decide which direction to go to.

    def findRandomPath(self,arraySizeX,arraySizeY,trail):

        self.currentNode = 0
        self.lastDirection = "Null"
        self.totalLength = 0
        self.uniqueNodes = []
        self.path = []
        #This while loop stops when the ant reaches the end of the matrix.

        while (self.currentNode != (arraySizeX*2-1)*(arraySizeY*2-1)-1):
            self.case = []
            specialCase = []
            self.trailSum = 0
            self.path.append(self.currentNode)
            #These 4 if statements are used to calculate the total difference for the current node to the adjacent nodes.
            #One if statement for a direction.
            #Only condition is that it cant go of the edge (out of bounds) of the matrix.
            #And cant go back to a node that the ant was previously on.

            #Since we are doing difference of node, there is a special case if the current node and the adjacent node is the same number.
            #This because 0/total will always give 0
            #Therefore we need to account for an option to go that node.

            if ((self.currentNode //((arraySizeX*2)-1)) < (arraySizeY*2 - 2) and self.lastDirection != "w") :
                if (self.calculateDifference("s",trail) != 0):
                    self.trailSum += self.returnPheromone("s",trail) * (1/abs(self.calculateDifference("s",trail)))
                    self.case.append("s")
                else:
                    specialCase.append("s")


            if (self.currentNode %(arraySizeX*2-1) < (arraySizeX*2-1) - 1 and self.lastDirection != "a") :
                if (self.calculateDifference("d",trail) != 0):
                    self.trailSum += self.returnPheromone("d",trail) * (1/abs(self.calculateDifference("d",trail)))
                    self.case.append("d")
                else:
                    specialCase.append("d")

            #For direction w (up) and a (left), when working with a small matrix, it is very unlikely that up and left are the right way to go.
            #Therefore there is a 50% the ant goes to that direction.
            #There is also an infinite loop that could occur if we did not have this.

            if (self.currentNode >=(arraySizeX*2-1) and self.lastDirection != "s") :
                if (self.calculateDifference("w",trail) != 0):
                    self.trailSum += self.returnPheromone("w",trail) * (1/abs(self.calculateDifference("w",trail)))
                    self.case.append("w")
                else:
                    rn = randint(0, 1)
                    if (rn == 1):
                        specialCase.append("w")


            if (self.currentNode % (arraySizeX*2-1) > 0 and self.lastDirection != "d") :
                if (self.calculateDifference("a",trail) != 0):
                    self.trailSum += self.returnPheromone("a",trail) * (1/abs(self.calculateDifference("a",trail)))
                    self.case.append("a")
                else:
                    rn = randint(0, 1)
                    if (rn == 1):
                        specialCase.append("a")
            


            #Here we are calculating the probablity of the ant going to an adjacent node using a number between 0 and 1
            #If there was no special case for the previous section, then the ant will go to nodes possible.
            #If there is a special node then, the ant will always go there.

            if len(specialCase) == 0:

                #These set of if statements are used if the ant has 1,2 or 3 nodes that it can go to respectively.
                #The probabilties for 3 paths are calculated by:
                #The number can either land between these.
                # 1/node1 -> 1/node2 + 1/node2 -> else

                if len(self.case) == 1:
                    self.moveAnt(self.case[0],arraySizeX,trail)

                elif len(self.case) == 2:
                    rn = random()
                    if (rn < (self.returnPheromone(self.case[0],trail) *(1/abs(self.calculateDifference(self.case[0],trail))/self.trailSum))):
                        self.moveAnt(self.case[0],arraySizeX,trail)
                    else:
                        self.moveAnt(self.case[1],arraySizeX,trail)

                elif len(self.case) == 3:
                    rn = random()
                    if (rn < self.returnPheromone(self.case[0],trail) *(1/abs(self.calculateDifference(self.case[0],trail))/self.trailSum)):
                        self.moveAnt(self.case[0],arraySizeX,trail)
                    elif (rn < ((self.returnPheromone(self.case[0],trail)*1/abs(self.calculateDifference(self.case[0],trail))+ \
                        (self.returnPheromone(self.case[0],trail)*1/abs(self.calculateDifference(self.case[1],trail))))/self.trailSum)):
                        
                        self.moveAnt(self.case[1],arraySizeX,trail)
                    else:
                        self.moveAnt(self.case[2],arraySizeX,trail)


            #For special cases, if there is only one, then we go to that.
            #Else we roll a number between the amount of special cases there are and chose the one we land on.

            elif len(specialCase) == 1:
                self.moveAnt(specialCase[0],arraySizeX,trail)
            elif len(specialCase) == 2:
                if (randint(0, 1) == 0):
                    self.moveAnt(specialCase[0],arraySizeX,trail)
                else:
                    self.moveAnt(specialCase[1],arraySizeX,trail)  
            else:
                randomnumber = randint(0, 2)
                if (randomnumber == 0):
                    self.moveAnt(specialCase[0],arraySizeX,trail)
                elif (randomnumber == 1):
                    self.moveAnt(specialCase[1],arraySizeX,trail)
                else:
                    self.moveAnt(specialCase[2],arraySizeX,trail)    
        self.path.append(self.currentNode)
        self.path.append((arraySizeX*2-1)*(arraySizeY*2-1)-1)
def runAntColony(numberOfAnts, numberOfCycles):
    

    trail = np.zeros((arraySizeY*2-1,arraySizeX*2-1))

    #Add one to every pheromone as a starter

    trail = trail + 1

    #Add nodes to our new matrix.

    for x in range(arraySizeX):
        for y in range(arraySizeY):
            trail[y*2][x*2] = randomArray[y][x]
    print(trail)

    #Here we are creating the ants and cycles.

    listOfAnts = []

    tic = timeit.default_timer()
    for i in range(numberOfAnts):
        listOfAnts.append(Ants())

    #This function adds decay to the problem

    def decayPhermone():
        for y in range(arraySizeY):
            for x in range(arraySizeX):
                if (x < arraySizeX-1) :
                    trail[y*2][(2*x)+1] *= 0.5
                if (y < arraySizeY-1) :
                    trail[y*2+1][2*x] *=0.5

    #This function runs all the ants.

    def cycle(number,trail):
        for cycle in range(number):
            decayPhermone()
            for ants in listOfAnts:
                ants.findRandomPath(arraySizeX,arraySizeY,trail)
            for ants in listOfAnts:
                for phermone in ants.uniqueNodes:
                    trail[phermone//(arraySizeX*2-1)][phermone % (arraySizeX*2-1)] += (1/ants.totalLength)

    cycle(numberOfCycles,trail)

    #This for loop finds the ant that travelled the shortest distance. 

    smallestLength = 99999
    for ants in listOfAnts:
        ants.findRandomPath(arraySizeX,arraySizeY,trail)
        if (ants.totalLength < smallestLength):
            smallestLength = ants.totalLength
            path = np.unique(ants.path)
    print(trail)

    #Prints out smallest length and computation time.

    print(smallestLength)
    toc = timeit.default_timer()
    print(toc - tic) 
    print(path)

    #This is not part of the algorithm, it is just to showcase the ant's path.

    for y1 in range(arraySizeY):
        print("")
        for x1 in range(arraySizeX):
            check = 0
            for scan in range(len(path)):
                if(path[scan] == x1*2+(y1*2*(arraySizeX*2-1))):
                    print("\033[91m" + str(trail[y1*2][x1*2].astype(int))  + "\033[0m" , end = ' ')
                    check = 1
            if(check == 0):
                print(str(trail[y1*2][x1*2].astype(int)) , end = ' ')
```
